```python
import pandas as pd
l1=[[] for i in range(30)]
for i in range(0,30):
    data=input()
    l1[i]=data.strip().split()
l2=[[] for i in range(30)]
p=0;
for i in range(0,30):
    k=min(l1[i])
    
    for j in l1[i]:
        if j==k:
            l1[i].remove(k)
            break
    for j in l1[i]:
        l2[p].append(j) 
           
    p+=1
```

```python
df=pd.DataFrame(l2,columns=['Name','Mark1','Mark2'])
```
